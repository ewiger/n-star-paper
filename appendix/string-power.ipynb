```python
from IPython.display import HTML, display

def display_table(data, highlight=False):
    html = "<table>"
    i = 0
    for row in data:
        html += "<tr>"
        j = 0
        for field in row:
            if field == "":
                field = "∅"
            html += f"<td><h4>{field}</h4></td>"
            j += 1
        html += "</tr>"
        i += 1
    html += "</table>"
    display(HTML(html))
```

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
display_table(data)
```

```python
def cdot(a, b):
    """Concatenation of two strings."""
    return f"{a}{b}"

def K(X=None):
    """Return a product table with string concatenation results over X."""
    if X is None:
        X = ["", "0", "1"]
    T = list()
    for i in range(len(X)):
        T.append(["" for k in range(len(X))])
        for j in range(len(X)):
            T[i][j] = cdot(X[i], X[j])
    return T

T = K()
display_table(T)
```

```python
def nice_list(X):
    X = list(set(X))
    X.sort()
    X = list(sorted(X, key=len))
    return X


def Splt(c, S):
    """
    String split operation
    
    Return a set of all strings that can be split from S by symbol c
    """
    if c == "":
        R = list()
        for i in range(len(S)):
            for j in range(len(S)):
                R.append(S[i:j+1])
    else:
        raise NotImplemented("Split over non-empty sub-strign is not implemented")
    return nice_list(R)
            
R = Splt("","0111")
R
```

```python
def flat_list(T):
    return nice_list([x for row in T for x in row]) 

def PwStr(X=None):
    """Power string"""
    X = flat_list(K(X))
    return X
        
X1 = PwStr()
X2 = PwStr(X1)

print(X1)
print(X2)

display_table(K(X1))
display_table(K(X2))
```
